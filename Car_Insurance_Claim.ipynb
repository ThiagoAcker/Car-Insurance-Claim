```python
#importing packages and reading the csv file
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

df = pd.read_csv(r'C:\Users\acker\GitHub\Car Insurance Claim\Car_Insurance_Claim.csv')
print(df.head())
print(df.columns.tolist())
print(df.info())
```

```python
#filling out the null data on 'CREDIT_SCORE' and 'ANNUAL_MILEAGE' columns
df['CREDIT_SCORE'].fillna(df['CREDIT_SCORE'].mean())
df['ANNUAL_MILEAGE'].fillna(df['ANNUAL_MILEAGE'].mean())
print(df.isnull().sum())
```

```python
#removing 'ID' and 'OUTCOME' columns
features = df.drop(columns=['ID', 'OUTCOME']).columns
#looping through the features
models = []
for col in features:
    model = logit(f'OUTCOME ~ {col}', data=df).fit()
    models.append(model)

#looping through models
accuracies = []
for feature in range(0, len(models)):
    conf_matrix = models[feature].pred_table()

    TN = conf_matrix[0, 0]
    FN = conf_matrix[1, 0]
    FP = conf_matrix[0, 1]
    TP = conf_matrix[1, 1]

    accuracy = (TP + TN)/(TN + TP + FN + FP)
    accuracies.append(accuracy)

#finding the feature with the largest accuracy
best_feature = features[accuracies.index(max(accuracies))]

#result
best_feature_df = pd.DataFrame({'best_feature': best_feature,
                                'best_accuracy': max(accuracies)}, index=[0])
best_feature_df
```

The analysis above identifies the feature "DRIVING_EXPERIENCE" as the greatest impactful on the predictive model performance, with an accuracy of 77%.
